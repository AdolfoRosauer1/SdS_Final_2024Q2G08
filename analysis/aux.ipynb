```python
import pandas as pd
import os
import json

# pick the 3 quickest realizations longer than 20 seconds
realizations = {}

def get_realizations(directory):
    if directory not in realizations:
        realizations[directory] = []
        
    for realization in os.listdir(directory):
        # read files that finish with {number}.csv
        # simulation_results/animaciones_finales/0,55/realization_0.55_1_vel.csv
        if realization.endswith("vel.csv"):
            continue
        df = pd.read_csv(os.path.join(directory, realization))
        if df["Time"].max() > 20:
            print(realization)
            realizations[directory].append(realization)

directory = "../simulation_results/animaciones_finales/0,55"
get_realizations(directory)
directory = "../simulation_results/animaciones_finales/0,7" 
get_realizations(directory)
directory = "../simulation_results/animaciones_finales/1,0"
get_realizations(directory)
directory = "../simulation_results/animaciones_finales/miti_miti"
get_realizations(directory)

# export realizations to a json file
with open("realizations.json", "w") as f:
    json.dump(realizations, f, indent=4)
```
